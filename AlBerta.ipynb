## RoBERTa

```python
from transformers import RobertaConfig, RobertaModel, RobertaTokenizerFast
```

```python
model = RobertaModel.from_pretrained('roberta-base')
```

## Check configuration og Reoberta Model

```python
model.config
```

```python
tokenizer = RobertaTokenizerFast.from_pretrained("roberta-base")
```

```python
input_ids = tokenizer.tokenize('It was a great day')
input_ids
```

* We can observe that the given sentence is tokenized, but what's that Ġ character? It is used
to indicate a space
* The RoBERTa tokenizer replaces all the white space with
the Ġ character.

```python
tokenizer.tokenize('I had a sudden epiphany')
```

* From the results, we can observe that the word epiphany is split into subwords **ep** and **iphany** since it is not present in the vocabulary

## Summary

* RoBERTa is a variant of BERT and it uses only the MLM task for training.
* Unlike BERT, it uses dynamic masking instead of static masking .
* It is trained with a large batch size. 
* It uses BBPE as a tokenizer and it has a vocabulary size of 50,000.

## SpanBert

Performing Q&As with pre-trained SpanBERT

```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline(
"question-answering",
model="mrm8488/spanbert-large-finetuned-squadv2",
tokenizer="SpanBERT/spanbert-large-cased"
)
```

```python
results = qa_pipeline({
'question': "What is machine learning?",
'context': "Machine learning is a subset of artificial intelligence. It is widely for creating a variety of applications such as email filtering and computer vision"
})
```

```python
print(results)
```

## Object Detection

```python
obj_detect = pipeline('object-detection',model='hustvl/yolos-base')
```

```python
!wget https://1.bp.blogspot.com/-2wyrfJjwJNw/Tpl2SKqQMkI/AAAAAAAADK8/I60vt30Rf6E/s1600/traffic+signal.jpg
```

```python
result = obj_detect('/content/traffic+signal.jpg')
```

```python
import torch
import torchvision
from torchvision.io import read_image
from torchvision.utils import draw_bounding_boxes
```

```python
result
```

```python
label = []
bbox = []
score= []
for item in result:
  print(item['score'])
  print(item['label'])
  print(item['box'])
  box = []
  for k,v in item['box'].items():
    box.append(v)
  print(box)
  label.append(item['label'])
  score.append(item['score'])
  bbox.append(box)
```

```python
bbox = torch.tensor(bbox, dtype=torch.int)
print(bbox)
print(bbox.size())
```

```python
%matplotlib inline
```

```python
img = read_image("/content/traffic+signal.jpg")
# draw bounding boxes on the input image
img=draw_bounding_boxes(img, bbox, width=5,labels= label,
                        font_size=500)
img = torchvision.transforms.ToPILImage()(img)
img.show()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(11,8),dpi=150)
plt.imshow(img)
```

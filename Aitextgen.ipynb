```python
from aitextgen import aitextgen

# Initialize the AI model
ai = aitextgen('EleutherAI/gpt-neo-1.3B')

# Set the prompt or starting text
prompt = "Write a blog about anti-money laundering."

# Generate the text
generated_text = ai.generate(n=1, prompt=prompt, return_as_list=True)

# Print the generated text
print(generated_text[0])
```
